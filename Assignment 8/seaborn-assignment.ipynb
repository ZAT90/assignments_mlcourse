```python
import seaborn as sns
import matplotlib.pyplot as plt
```

**Do the following tasks:** 
* set style to '`white`'
* load titanic dataset into '`titanic`', tips into '`tips`' and iris into '`iris`'

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
sns.set_style('white')

titanic = sns.load_dataset('titanic')

titanic.head()
```

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
iris = sns.load_dataset('iris')
iris.head()
```

**Recreate the plot below. Use tips dataset. Hint: pay attention to x and y labels.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
plt.figure(figsize=(4,4))
custom_colors = {"Male":"#5975a4","Female":"#5f9e6e"}
outlier_style = dict(markerfacecolor='black', marker='d', markersize=3, linestyle='none')
sns.boxplot(x='day', y='tip', data=tips,hue='sex',palette=custom_colors,flierprops=outlier_style)
# plt.title("Age distribution: Survived vs not Survived")
plt.ylabel("tip")
plt.tight_layout()
plt.show()
```

**Recreate the plot below. Use tips dataset. Hint: pay attention to x and y labels.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!

jplot = sns.jointplot(data=tips,x='total_bill',y='tip',kind='hex')

plt.show()
```

**Recreate the distplot given below. Use titanic dataset. Hint: pay attention to x and y labels.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
# titanic.head()
plt.figure(figsize=(6,4))
sns.histplot(data=titanic,x='fare',color='green',alpha=0.3,bins=20,edgecolor='none')

plt.show()
```

**Recreate the boxplot below using titanic dataset. Hint: pay attention to x and y labels.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
plt.figure(figsize=(5,5))
custom_colors = {"no":"#b5caf3","yes":"#ecbca7"}
outlier_style = dict(markerfacecolor='black', marker='d', markersize=3, linestyle='none')
sns.boxplot(x='class', y='age', data=titanic,hue='alive',
             palette=custom_colors,flierprops=outlier_style
            )
# plt.title("Age distribution: Survived vs not Survived")
plt.ylabel("age")
plt.tight_layout()
plt.show()
```

**Recreate the plot below using titanic dataset. Hint: pay attention to x and y labels.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
jplot = sns.jointplot(data=titanic,x='fare',y='age')

plt.show()
```

**Recreate the plot below. Use titanic dataset. ** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
plt.figure(figsize=(5,4))
diff_colors = ['#1ea5ce','#90efb7','#ebb274']
custom_colors = {"first":"blue","second":"green","third":"orange"}
sns.boxplot(x='class',y='age',hue='class',data=titanic,flierprops=outlier_style, 
             palette=diff_colors, 
            legend=False)
plt.ylabel('age')
plt.show()
```

**Can you plot how many male and female were on titanic? Separate them based on their class.**

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
plt.figure(figsize=(5,3))
diff_colors = ['#5975a4','#5f9e6e','#bd5d60']
sns.countplot(data=titanic,x='sex',hue='class',palette=diff_colors,linewidth=0)
plt.show()
```

**Recreate the plot below. Use titanic dataset. Hint: Matrix plots** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
plt.figure(figsize=(8,4))
sns.heatmap(titanic.corr(numeric_only=True),annot=True,cmap='coolwarm')
plt.title('titanic.corr()')
plt.yticks(rotation=0)
plt.show()
```
