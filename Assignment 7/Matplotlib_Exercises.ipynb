In the first cell, we are simply creating data x and y to use in the provided tasks.

```python
# We need data to work with, lets create data first!
import numpy as np
x = np.arange(0,10)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
# plt.show() for non-notebook users
plt.show()
```

**1. Instructions:**
* ** Use <code>plt.figure()</code> and create a figure object, say fig.**
* ** Add an axis to the figure canvas (`fig`), `[left, bottom, width, height] is [0,0,1,1]`. <i>Hint: <code>add_axes()</code><i>**
* ** Plot the data x,y on the axes**
* ** Set the labels x and y**
* ** Set the title**
* ** Create legends as shown in the plot**

Match the output below: 

```python
#Code here
fig = plt.figure(figsize=(5,3))

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,'g-',label='y = x*2')
# Adding the text box

axes.set_xlabel("x_label")
axes.set_ylabel("y_label")
axes.set_title("Plot_title")
axes.legend(loc='lower right', frameon=True, facecolor='white', edgecolor='black')
plt.show()
```

**2. Instructions**<br>
* ** Create a figure object**
* ** Put two axes on the figure at [0,0,1,1] and [0.1,0.6,0.2,0.2]**
* ** Plot data (x,z) on main axes**
* ** Plot data (x,z) on the small axes and set its x limit b/w 1, 2, y limit 0, 5**
* ** Replicate the figure to axes, titles, legends etc**

```python
#Code here
fig = plt.figure(figsize=(6,3))
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.1,0.6,0.2,0.2])
axes1.plot(x,z,'r-',label='y_main = x_main*2')

axes1.set_xlabel("x_main")
axes1.set_ylabel("y_main")
axes1.set_title("Title_main")
axes1.legend(loc='lower right', frameon=True, facecolor='white', edgecolor='grey')

# Set the limits for the inset
axes2.plot(x,z,'b-')
axes2.set_xlim([1, 2])
axes2.set_ylim([0, 5])
axes2.set_xlabel("x_inset")
axes2.set_ylabel("y_inset")
axes2.set_title("inset")

# plt.tight_layout()
plt.show()
```
