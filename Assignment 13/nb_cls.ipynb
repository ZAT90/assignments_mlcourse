```python
import pandas as pd
```

```python
df = pd.read_csv("fake_or_real_news.csv")
df
```

```python
df.info()
```

```python
df_featured = df[['text','label']].dropna(subset=['text', 'label'])
df_featured.info()
```

```python
X = df_featured.drop('label',axis=1)
Y = df_featured['label']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.20,random_state=42)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectortizer = CountVectorizer()
```

```python
X_train_vectorized = vectortizer.fit_transform(X_train['text'])
X_test_vectorized = vectortizer.transform(X_test['text'])
```

```python
model.fit(X_train_vectorized,Y_train)
print(X_train_vectorized.shape)
print(X_test_vectorized.shape)
```

```python
# Y_pred = model.predict(X_test_vectorized)
# from sklearn.metrics import accuracy_score
print(model.score(X_test_vectorized, Y_test))
```

In this task, the fake_or_real_news dataset was loaded and explored using pandas. The text and label columns were selected because text is the input feature and label is the target variable. Missing values were removed using dropna() to avoid errors during text vectorization.

The dataset was then split into training and testing sets. Since Naive Bayes cannot directly work with raw text, CountVectorizer was used to convert the news text into numerical features. A Multinomial Naive Bayes model was trained using the vectorized training data.

Finally, the model was tested using the test data, and accuracy_score was used to evaluate the model. The accuracy shows how many news articles were correctly classified as FAKE or REAL.
